# Probability

For this project, I am playing in a single deck blackjack game, and I have seen 4 hands against the dealer . In all 4 of those hands, no ace or 10 has appeared, I have seen a total of 24 cards.

My probability of getting a **Ace** ***A*** as a first card is  **$\frac{4}{28}$** 

My probability of getting an **10** ***T*** as a first card is $\frac{4}{28}$

My probaility of getting neither an ace and 10 ***N*** is $\frac{5}{7}$


banluck-1280x720.jpeg


```python
p = [4/28,4/28,5/7]
```

```python
sum(p) 
```

I am going to create a compund event,In this case, compund event is getting either Ace or 10 . I am going to call this draw **S**


```python
ProbabilityS = 0

for i in [0,2]:
  ProbabilityS += p[i]
ProbabilityS
```

As per the requirements, I am going to calculate the probabiity of union and intersection of these events.

The probability of getting Ace and either Ace or 10 in two draws

$$P(A\cap S)$$
 

```python
intersection = p[0]*ProbabilityS
intersection
```

The probability of getting ace in second draw given that you  got either Ace or 10 is first draw
$$
P(A|S)=\frac{P(A\cap S)}{P(S)}
$$

```python
SGA = intersection/ProbabilityS
```

```python
SGA
```

The probability of getting Ace or 10 in second draw given that you got Ace on first draw:
$$
P(S|A)=\frac{P(S\cap A)}{P(A)}
$$

```python
AGS = intersection/p[0]
```

```python
AGS
```

ApplyinG Apple Baye's Theorem
$$
P(A|S) = \frac{P(A)P(S|A)}{P(S)}
$$

```python
SGA == (p[0]* AGS)/ProbabilityS
```

For the simulation part, I will be computing the frequency counts so, i will be multiplying by 28.




```python
p28 = [28*i for i in p]
p28
```

I am going to simulate a simple occurance of my sample space 10 times using random library.

```python
from random import choices


draw = choices(['Ace','Ten','Other'],p28,k=10)
print(draw)

countAce = 0;
countTen = 0;
countOthers = 0;

for i in draw:
  if i == 'Ace':
    countAce += 1
  if i == 'Ten':
    countTen += 1
  if i == 'Other':
    countOthers += 1

estimateProb = [countAce/len(draw), countTen/len(draw), countOthers/len(draw)]
estimateProb
```
