# **Function**

For this project, I have created a function with at least three features.

 $$

```python
import math
import numpy as np
#creating a function namee rato and it can only take one parameter
def rato(x):
  return (1 + ((x**2-2)/(math.pi*np.e*x)))
```

```python
#calling a function   and passing  an argument 
rato(2)
```
